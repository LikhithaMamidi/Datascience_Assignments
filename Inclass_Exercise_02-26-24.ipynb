### calculate covariance matrix for iris dataset and (petal length, petal width)

```python
iris = load_iris()
iris_data = iris.data
petal_data = iris_data[:, 2:4]
covariance_matrix = np.cov(petal_data, rowvar=False)

print("Covariance Matrix for Petal Length and Petal Width:\n", covariance_matrix)
```

## Discuss the results and what the covariance values indicate about the variable relationship 
* The covariance matrix will be a 2x2 matrix. The diagonal elements represent the variance of each variable, and the off-diagonal elements represent the covariance between the variables.
* Positive covariance indicates that as one variable increases, the other tends to increase.

### calculate covariance matrix for iris dataset and sepal length ,sepal width

```python
sepal_data = iris_data[:, :2]

covariance_matrix_sepal = np.cov(sepal_data, rowvar=False)

print("Covariance Matrix for Sepal Length and Sepal Width:\n", covariance_matrix_sepal)
```

## Discuss the results and what the covariance values indicate about the variable relationship 
* Similar to Sepal, the covariance matrix for Petal will also be a 2x2 matrix.
* A positive covariance suggests a positive relationship, and a negative covariance suggests a negative relationship.
* The magnitude of covariance indicates the strength of the relationship. Larger magnitudes suggest stronger relationships.

### include the correlation coefficients on scatter plots for better interpretations

```python
iris = load_iris()
iris_df = sns.load_dataset('iris')
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sepal_plot = sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris_df)
corr_sepal = iris_df[['sepal_length', 'sepal_width']].corr().iloc[0, 1]
plt.title(f'Sepal Length vs Sepal Width\nCorrelation Coefficient: {corr_sepal:.2f}')
plt.subplot(1, 2, 2)
petal_plot = sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=iris_df)
corr_petal = iris_df[['petal_length', 'petal_width']].corr().iloc[0, 1]
plt.title(f'Petal Length vs Petal Width\nCorrelation Coefficient: {corr_petal:.2f}')

plt.tight_layout()
plt.show()
```
